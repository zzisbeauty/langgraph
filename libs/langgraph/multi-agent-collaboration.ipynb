````python
# ÂÆö‰πâÂ∑•ÂÖ∑
from typing import Annotated
from langchain_tavily import TavilySearch
from langchain_core.tools import tool


from langchain_experimental.utilities import PythonREPL

repl = PythonREPL()


#  =========#  =========#  =========#  =========#  ========= ÂàõÂª∫ TOOLS


""" ‚ÄúÂ§ö Agent Âçè‰Ωú‚ÄùÔºàMulti-Agent CollaborationÔºâÊû∂ÊûÑ

# ÂÆÉÂÆûÁé∞‰∫Ü‰∏Ä‰∏™Âèå Agent Âæ™ÁéØÁ≥ªÁªüÔºö‰∏Ä‰∏™Ë¥üË¥£ÊêúÁ¥¢ÔºàResearcherÔºâÔºå‰∏Ä‰∏™Ë¥üË¥£ÁªòÂõæÔºàChart GeneratorÔºâÔºå‰∏§ËÄÖ‰∫§ÊõøÂ∑•‰ΩúÁõ¥Âà∞ÂæóÂá∫ÊúÄÁªàÁ≠îÊ°à„ÄÇ
"""

# tools -1 
tavily_tool = TavilySearch(max_results=5)


# tools - 2

@tool
def python_repl_tool(code: Annotated[str, "Áî®‰∫éÁîüÊàê‰Ω†ÁöÑÂõæË°®Âπ∂ÊâßË°åÁöÑ Python ‰ª£Á†Å"],):
    """‰ΩøÁî®Ê≠§Â∑•ÂÖ∑Êù•ÊâßË°å Python ‰ª£Á†Å„ÄÇÂ¶ÇÊûú‰Ω†ÊÉ≥Êü•ÁúãÊüê‰∏™ÂÄºÁöÑËæìÂá∫ÔºåÂ∫îÂΩì‰ΩøÁî® print(...) Â∞ÜÂÖ∂ÊâìÂç∞Âá∫Êù•„ÄÇËøôÊ†∑Áî®Êà∑Â∞±ÂèØ‰ª•ÁúãÂà∞ÁªìÊûú„ÄÇ"""
    try:
        result = repl.run(code)
    except BaseException as e:
        return f"Failed to execute. Error: {repr(e)}"
    result_str = f"Successfully executed:\n```python\n{code}\n```\nStdout: {result}"
    return (result_str + "\n\nIf you have completed all tasks, respond with FINAL ANSWER.")


# @tool  
# def python_repl_tool(code: Annotated[str, "The python code to execute to generate your chart."]):  
#     """Use this to execute python code. If you want to see the output of a value,  
#     you should print it out with `print(...)`. This is visible to the user."""  
#     try:  
#         # Ê∑ªÂä† matplotlib ÈÖçÁΩÆ  
#         import matplotlib  
#         matplotlib.use('Agg')  # ‰ΩøÁî®Èùû‰∫§‰∫íÂºèÂêéÁ´Ø  
#         import matplotlib.pyplot as plt  
          
#         result = repl.run(code)  
          
#         # Â¶ÇÊûú‰ª£Á†Å‰∏≠ÂàõÂª∫‰∫ÜÂõæË°®Ôºå‰øùÂ≠òÂπ∂ÊòæÁ§∫  
#         if 'plt' in code or 'matplotlib' in code:  
#             plt.savefig('/tmp/chart.png', dpi=150, bbox_inches='tight')  
#             plt.close()  
#             return f"Successfully executed:\n```python\n{code}\n```\nStdout: {result}\n\nChart saved to /tmp/chart.png"  
          
#     except BaseException as e:  
#         return f"Failed to execute. Error: {repr(e)}"  
      
#     return (  
#         result + "\n\nIf you have completed all tasks, respond with FINAL ANSWER."  
#     )


# # ÂçïÁã¨ÊµãËØï PythonREPL ÂõæË°®ÁîüÊàê  
# test_code = """  
# import matplotlib.pyplot as plt  
# years = [2019, 2020, 2021, 2022, 2023]  
# gdp = [2.83, 2.76, 3.13, 3.07, 3.13]  
  
# plt.figure(figsize=(10, 6))  
# plt.plot(years, gdp, marker='o')  
# plt.title('UK GDP (2019-2023)')  
# plt.savefig('test_chart.png')  
# print("Chart saved as test_chart.png")  
# """  
# result = python_repl_tool.invoke(test_code)  
# print(result)



# ==================# ==================# ==================# ==================# ==================

from typing import Literal
from langgraph.types import Command
from langchain_core.messages import BaseMessage, HumanMessage
from langgraph.prebuilt import create_react_agent
from langgraph.graph import MessagesState, END




# ========# ========# ========# ========# ========# ========# ========  ÂàõÂª∫  Agent  ÂâçÁöÑÂáÜÂ§á



# Agent Ë∞ÉÁî®ÂæóÂà∞ÂÖ∂ÈúÄË¶ÅÁöÑÁ≥ªÁªüÊèêÁ§∫ -  Âçè‰ΩúÂçèËÆÆ - Á≥ªÁªüÁ∫ßÊèêÁ§∫
def make_system_prompt(suffix: str) -> str:
    return (
        """
        ‰Ω†ÊòØ‰∏ÄÂêçÊúâÂ∏ÆÂä©ÁöÑ AI Âä©ÊâãÔºå‰∏éÂÖ∂‰ªñÂä©ÊâãÂçè‰Ωú„ÄÇ
        ‰ΩøÁî®ÊâÄÊèê‰æõÁöÑÂ∑•ÂÖ∑Êù•Êé®ËøõÈóÆÈ¢òÁöÑËß£Á≠î„ÄÇ
        Â¶ÇÊûú‰Ω†Êó†Ê≥ïÂÆåÂÖ®ÂõûÁ≠îÔºå‰πüÊ≤°ÂÖ≥Á≥ªÔºåÂè¶‰∏Ä‰ΩçÊã•Êúâ‰∏çÂêåÂ∑•ÂÖ∑ÁöÑÂä©Êâã
        ‰ºöÂú®‰Ω†ÂÅú‰∏ãÁöÑÂú∞ÊñπÁªßÁª≠Â∏ÆÂä©„ÄÇËØ∑Â∞Ω‰Ω†ÊâÄËÉΩÊâßË°åÂπ∂ÂèñÂæóËøõÂ±ï„ÄÇ
        """
        f"\n{suffix}"
    )

# Â¶ÇÊûú‰Ω†ÊàñÂÖ∂‰ªñ‰ªª‰ΩïÂä©ÊâãÂ∑≤ÁªèÂæóÂá∫ÊúÄÁªàÁ≠îÊ°àÊàñÂèØ‰∫§‰ªòÊàêÊûúÔºå
#         ËØ∑Âú®‰Ω†ÁöÑÂõûÂ§çÂâçÂä†‰∏ä FINAL ANSWERÔºå‰ª•‰æøÂõ¢ÈòüÁü•ÈÅìÂèØ‰ª•ÂÅúÊ≠¢„ÄÇ


# ========# ========# ========# ========# ========# ========# ========  ÂàõÂª∫  Agent


# Research agent and node =========== Agent 1
# ÊñπÊ≥ïËß£Èáä https://share.google/aimode/C942mfmtaA5H8atE3
# Ëøô‰∏™ Agent Âíå node ‰πãÈó¥ÊòØ‰ªÄ‰πàÂÖ≥Á≥ª„ÄÇÂÆÉÈáåÈù¢Êúâ node Ëøô‰∏™Ê¶ÇÂøµ‰ª£Ë°®ÁöÑÂäüËÉΩÊ®°ÂùóÔºü  https://share.google/aimode/Ec9lXcc46uAEzJzlH
research_agent = create_react_agent(
    llm,
    tools=[tavily_tool], # ÁªëÂÆöÂ∑•ÂÖ∑
    prompt=make_system_prompt(
        "You can only do research. You are working with a chart generator colleague. "  
        "IMPORTANT: Never include 'FINAL ANSWER' in your response. "  
        "Always let the chart generator create the visualization."  
    ), # ËäÇÁÇπÁ∫ßÊèêÁ§∫ÔºàAgent ‰∏ìÁî® promptÔºâ
)


# Chart generator agent and node =========== Agent 2
# NOTE: THIS PERFORMS ARBITRARY CODE EXECUTION, WHICH CAN BE UNSAFE WHEN NOT SANDBOXED
# Agent Ëß£ÈáäËØ¥ÊòéÔºå‰∏ªË¶ÅÊòØ tool ÁöÑÁªëÂÆöÊúâÁñëÈóÆÔºö https://share.google/aimode/9eXTN36FySagtEYVm
# Ëøô‰∏™AgentÁöÑÊñπÊ≥ïÁöÑËØ¶ÁªÜË∞ÉÁî®ËøáÁ®ãÔºö https://share.google/aimode/YtjLaSoDRBhWBrFH8   ÂæàÈáçË¶ÅÔºåÂæàÊ∏ÖÊ•öÁöÑËØ¥Êòé  Tool ÂØπ‰∫é Agent ÁöÑÊ¶ÇÂøµ
chart_agent = create_react_agent(
    llm,
    [python_repl_tool],
    prompt=make_system_prompt("‰Ω†Ë¥üË¥£ÁîüÊàêÂõæË°®„ÄÇ‰Ω†Ê≠£Âú®‰∏é‰∏Ä‰ΩçË¥üË¥£Á†îÁ©∂ÁöÑÂêå‰∫ãÂçè‰Ωú„ÄÇ"),
)


# ‰∏äËø∞‰∏§‰∏™ Agent ÁöÑ tool ÁöÑÂå∫Âà´Ôºö https://share.google/aimode/XMpjXwINTFo0o5DKg
````

````python
# Áî®‰∫éÂÆûÁé∞Á†îÁ©∂Êô∫ËÉΩ‰ΩìÂíåÂõæË°®ÁîüÊàêÊô∫ËÉΩ‰Ωì‰πãÈó¥ÁöÑÂçè‰Ωú

# Ë∑ØÁî±ÈÄªËæëÂáΩÊï∞
def get_next_node(last_message: BaseMessage, goto: str):
    if "FINAL ANSWER" in last_message.content:
        # Any agent decided the work is done
        return END
    return goto


def research_node(state: MessagesState,) -> Command[Literal["chart_generator", END]]:
    result = research_agent.invoke(state)
    goto = get_next_node(result["messages"][-1], "chart_generator")
    # wrap in a human message, as not all providers allow
    # AI message at the last position of the input messages list
    result["messages"][-1] = HumanMessage(
        content=result["messages"][-1].content, name="researcher"
    )
    return Command(  # ËøîÂõû Command ÂØπË±°ÔºåÂåÖÂê´Áä∂ÊÄÅÊõ¥Êñ∞ÂíåË∑ØÁî±‰ø°ÊÅØ
        update={
            # share internal message history of research agent with other agents
            "messages": result["messages"],
        },
        goto=goto,
    )


def chart_node(state: MessagesState) -> Command[Literal["researcher", END]]:
    result = chart_agent.invoke(state)
    goto = get_next_node(result["messages"][-1], "researcher")
    # wrap in a human message, as not all providers allow
    # AI message at the last position of the input messages list
    result["messages"][-1] = HumanMessage(
        content=result["messages"][-1].content, name="chart_generator"
    )
    return Command(
        update={
            # share internal message history of chart agent with other agents
            "messages": result["messages"],
        },
        goto=goto,
    )
````

````python
# ÂÆö‰πâÂõæÁªìÊûÑÔºåÂπ∂Êú™ÊâßË°å

from langgraph.graph  import StateGraph,  START

workflow = StateGraph(MessagesState)
workflow.add_node("researcher", research_node)
workflow.add_node("chart_generator", chart_node)

# START ÊòØ‰∏Ä‰∏™ÁâπÊÆäÁöÑËôöÊãüËäÇÁÇπÔºåÁî®‰∫éÂÆö‰πâÂõæÁöÑÂÖ•Âè£ÁÇπ„ÄÇÂΩì‰Ω†Ê∑ªÂä†‰ªé START Âà∞Êüê‰∏™ËäÇÁÇπÁöÑËæπÊó∂Ôºå‰Ω†ÊòØÂú®ÂëäËØâÂõæÊâßË°åÂºïÊìéÔºö"ÂΩìÂõæÂºÄÂßãÊâßË°åÊó∂ÔºåÈ¶ñÂÖàËøêË°åËøô‰∏™ËäÇÁÇπ"„ÄÇ
workflow.add_edge(START, "researcher") #  ÂõæÁªìÊûÑ‰ªéËøôÈáåÂºÄÂßã
# graph = workflow.compile() # # ÁºñËØëÂõæÔºå‰ΩÜËøòÊ≤°ÊúâÊâßË°å
graph = workflow.compile(  
    interrupt_before=["chart_generator"],  # Ë∞ÉËØïÊñ≠ÁÇπ  
    interrupt_after=["researcher"]         # Ë∞ÉËØïÊñ≠ÁÇπ  
)
# ÁºñËØëÂõæÂêéÔºåÂõæÁªìÊûÑ‰∏≠ Agent Â∞Ü‰ºöÊ†πÊçÆ State ÂÜ≥ÂÆöËµ∞ÂêëÂì™‰∏™ node ËäÇÁÇπÔºö
# https://deepwiki.com/search/-def-researchnodestate-message_8f0ae874-d6d3-49d2-a31a-d632ece50063?mode=fast#10
````
